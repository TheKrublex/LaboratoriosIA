```python
from treelib import Node, Tree

# Salida: arbol en base al laberinto
def createMaze():
  tree = Tree()
  # Root
  tree.create_node("e9", "e9", data = 0)
  # Rama 1
  tree.create_node("a9", "a9", parent="e9", data = 4)
  tree.create_node("e7", "e7", parent="e9", data = 2)
  tree.create_node("g7", "g7", parent="e9", data = 4)
  # Rama 2
  tree.create_node("c5", "c5", parent="e7", data = 8)
  tree.create_node("f5", "f5", parent="e7", data = 3)
  # Rama 3
  tree.create_node("f3", "f3", parent="f5", data = 2)
  tree.create_node("i5", "i5", parent="f5", data = 3)
  # Rama 4
  tree.create_node("h1", "h1", parent="f3", data = 4)
  tree.create_node("c3", "c3", parent="f3", data = 3)
  tree.create_node("j1", "j1", parent="f3", data = 6)
  tree.create_node("i7", "i7", parent="i5", data = 2)
  tree.create_node("j5", "j5", parent="i5", data = 1)
  # Rama 5
  tree.create_node("b3", "b3", parent="c3", data = 1)
  tree.create_node("c1", "c1", parent="c3", data = 2)
  tree.create_node("j7", "j7", parent="i7", data = 1)
  tree.create_node("j9", "j9", parent="i7", data = 3)
  # Rama 6
  tree.create_node("d1", "d1", parent="c1", data = 1)
  tree.create_node("b1", "b1", parent="c1", data = 1)
  return tree

# Entrada: arbol, costo de visitar cada casilla del laberinto
# Salida: diccionario con -> Pasos, Nodos Visitados, Costo, Si encuentra o no la ruta 
def depthFirst(tree, cost):
  steps = 0
  visitedNodes = []
  visitedNodesCost = []
  visitedNodesTemp = []
  index = 0
  totalCost = 0
  path = False
  # Recorrido
  for node in tree.expand_tree(mode=1):
    steps += tree.get_node(node).data
    totalCost += tree.get_node(node).data
    if node not in visitedNodes:
        visitedNodes.append(node)

    #Costo Ruta Directa
    visitedNodesTemp = visitedNodesCost
    for i in visitedNodesTemp:
      index += 1
      if tree.level(node) == tree.level(i):
        for x in visitedNodesCost[index-1:]:
          steps += tree.get_node(x).data
          totalCost -= tree.get_node(node).data
        visitedNodesCost = visitedNodesCost[:index-1]
        break
    visitedNodesCost.append(node)
    index = 0
    #

    if node == "h1":
      path = True
      break
  #totalCost = len(visitedNodesCost) - 1 
  dic = {"steps": steps, "visitedNodes": visitedNodes, "visitedNodesCost": visitedNodesCost, "totalCost": totalCost * cost, "path": path}
  return dic

# Entrada: arbol, costo de visitar cada casilla del laberinto
# Salida: diccionario con -> Pasos, Nodos Visitados, Costo, Si encuentra o no la ruta 
def breadthFirst(tree, cost):
  steps = 1
  visitedNodes = []
  visitedNodesCost = []
  totalCost = 0
  path = False
  # Recorrido
  for node in tree.expand_tree(mode=2):
    if node not in visitedNodes:
        visitedNodes.append(node)
    if node == "h1":
      path = True
      break
  #obtener la ruta
  if path:
    tempNode = tree.get_node("h1")
    while tempNode != None:
      visitedNodesCost.insert(0, tempNode.identifier)
      totalCost += tempNode.data
      tempNode = tree.parent(tempNode.identifier)
  #obtener los pasos
  for node in visitedNodes[::-1]:
    tempNode = tree.get_node(node)
    if(tempNode.identifier == "e9"):
        break
    elif(tempNode.identifier == "h1"):
      steps += tempNode.data
    elif(tempNode.is_leaf()):
      steps += tempNode.data * 2
    else:
      hasChildren = False
      for n in tree.children(tempNode.identifier):
        if n in visitedNodes:
          hasChildren = True
          break
      if not hasChildren:
        steps += tempNode.data * 2
      else:
        if(visitedNodes[visitedNodes.index(tempNode.identifier) + 1] in tree.siblings(tempNode.identifier)):
          steps += tempNode.data * 3
        else:
          steps += tempNode.data
  dic = {"steps": steps, "visitedNodes": visitedNodes, "visitedNodesCost": visitedNodesCost, "totalCost": totalCost * cost, "path": path}
  return dic

# Entrada: 2 diccionarios, cada uno con los resultados de los algoritmos
def printRes(depth, breadth):
  print("DEPTH FIRST")
  print("Pasos: " + str(depth["steps"]))
  print("Nodos visitados: " + str(len(depth["visitedNodes"])) + " -> " + str(depth["visitedNodes"]))
  print("Nodos visitados Ruta: " + str(len(depth["visitedNodesCost"])) + " -> " + str(depth["visitedNodesCost"]))
  print("Costo Ruta: " + str(depth["totalCost"]))
  print("Se encontró la Ruta?: " + str(depth["path"]) + "\n")
  print("BREADTH FIRST")
  print("Pasos: " + str(breadth["steps"]))
  print("Nodos visitados: " + str(len(breadth["visitedNodes"])) + " -> " + str(breadth["visitedNodes"]))
  print("Nodos visitados Ruta: " + str(len(breadth["visitedNodesCost"])) + " -> " + str(breadth["visitedNodesCost"]))
  print("Costo Ruta: " + str(breadth["totalCost"]))
  print("Se encontró la Ruta?: " + str(breadth["path"]))
```

```python
def main(cost = 1):
  tree = createMaze()
  printRes(depthFirst(tree, cost), breadthFirst(tree, cost))

main()

# CONCLUSIONES

# DEPTH FIRST: Este algoritmo a pesar de recorrer mas nodos en comparacion al
#              algoritmo de anchura es mas eficiente, esto se ve reflejado en la cantidad de pasos
#              que se deben recorrer desde el origen hasta el destino, sin embargo,
#              esto es solo en este caso en especifico ya que el nodo destino está
#              en los ultimos niveles a la izquierda, y este algoritmo funciona
#              recorriendo de izquierda a derecha siempre buscando los nodos que están
#              mas al fondo; si el destino se hubiera encontrado al fondo pero a la derecha
#              y se utiliza el mismo recorrido probablemente el algoritmo de anchura seria mas eficiente.
#              Tambien encontrar la ruta en este algoritmo fue mas facil ya que conforme se ejecutaba el 
#              algoritmo se podian ir descartando los nodos que no eran parte del recorrido, esto eliminando
#              los nodos anteriores que se encontraban en el mismo nivel del nodo actual.

# BREADTH FIRST: Este algoritmo muestra buenos resultados, sin embargo, para este caso en especifico
#                el algoritmo no es el mas util, ya que encontrar cual es la ruta encontrada por el
#                algoritmo es mas complejo y no se pueden eliminar los nodos que no son parte de la solucion
#                porque hasta no encontrar el nodo destino no se saben cuales nodos son los que siguen el 
#                "camino" hasta el destino, por lo que para encontrar la solucion primero se encontr el nodo
#                destino y despues se hizo un recorrido en reversa desde el nodo destino escogiendo cada uno
#                de sus padres hasta llegar al nodo origen. Igualmente debido a que se recorren todos los nodos
#                de un nivel, se realizan muchos pasos e incluso un solo nodo se tiene que visitar hasta 3 veces 
#                para continuar en busca de su destino. Es cierto que se recorrieron menos nodos, pero en el caso
#                del laberinto se recorrieron pasos que fueron innecesarios.
```
